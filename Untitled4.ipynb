```python
from sklearn.preprocessing import LabelEncoder

# Encode 'N'
label_n = LabelEncoder()
data['Encoded_N'] = label_n.fit_transform(data['N'])

# Encode 'P'
label_p = LabelEncoder()
data['Encoded_P'] = label_p.fit_transform(data['P'])

# Encode 'K'
label_k = LabelEncoder()
data['Encoded_K'] = label_k.fit_transform(data['K'])

# Encode 'temperature'
label_temperature = LabelEncoder()
data['Encoded_Temperature'] = label_temperature.fit_transform(data['temperature'])

# Encode 'humidity'
label_humidity = LabelEncoder()
data['Encoded_Humidity'] = label_humidity.fit_transform(data['humidity'])

# Encode 'ph'
label_ph = LabelEncoder()
data['Encoded_pH'] = label_ph.fit_transform(data['ph'])

# Encode 'rainfall'
label_rainfall = LabelEncoder()
data['Encoded_Rainfall'] = label_rainfall.fit_transform(data['rainfall'])

# Create mapping dictionaries for 'N'
mapping_n = dict(zip(data['Encoded_N'].unique(), label_n.inverse_transform(data['Encoded_N'].unique())))
sorted_mapping_n = sorted(mapping_n.items(), key=lambda x: x[0])

# Create mapping dictionaries for 'P'
mapping_p = dict(zip(data['Encoded_P'].unique(), label_p.inverse_transform(data['Encoded_P'].unique())))
sorted_mapping_p = sorted(mapping_p.items(), key=lambda x: x[0])

# Create mapping dictionaries for 'K'
mapping_k = dict(zip(data['Encoded_K'].unique(), label_k.inverse_transform(data['Encoded_K'].unique())))
sorted_mapping_k = sorted(mapping_k.items(), key=lambda x: x[0])

# Create mapping dictionaries for 'temperature'
mapping_temperature = dict(zip(data['Encoded_Temperature'].unique(), label_temperature.inverse_transform(data['Encoded_Temperature'].unique())))
sorted_mapping_temperature = sorted(mapping_temperature.items(), key=lambda x: x[0])

# Create mapping dictionaries for 'humidity'
mapping_humidity = dict(zip(data['Encoded_Humidity'].unique(), label_humidity.inverse_transform(data['Encoded_Humidity'].unique())))
sorted_mapping_humidity = sorted(mapping_humidity.items(), key=lambda x: x[0])

# Create mapping dictionaries for 'ph'
mapping_ph = dict(zip(data['Encoded_pH'].unique(), label_ph.inverse_transform(data['Encoded_pH'].unique())))
sorted_mapping_ph = sorted(mapping_ph.items(), key=lambda x: x[0])

# Create mapping dictionaries for 'rainfall'
mapping_rainfall = dict(zip(data['Encoded_Rainfall'].unique(), label_rainfall.inverse_transform(data['Encoded_Rainfall'].unique())))
sorted_mapping_rainfall = sorted(mapping_rainfall.items(), key=lambda x: x[0])

# Print the mapping dictionaries for each encoded feature
print("Encoded_N Mapping:")
for key, value in sorted_mapping_n:
    print(f"{key} = {value}")

print("\nEncoded_P Mapping:")
for key, value in sorted_mapping_p:
    print(f"{key} = {value}")

print("\nEncoded_K Mapping:")
for key, value in sorted_mapping_k:
    print(f"{key} = {value}")

print("\nEncoded_Temperature Mapping:")
for key, value in sorted_mapping_temperature:
    print(f"{key} = {value}")

print("\nEncoded_Humidity Mapping:")
for key, value in sorted_mapping_humidity:
    print(f"{key} = {value}")

print("\nEncoded_pH Mapping:")
for key, value in sorted_mapping_ph:
    print(f"{key} = {value}")

print("\nEncoded_Rainfall Mapping:")
for key, value in sorted_mapping_rainfall:
    print(f"{key} = {value}")
```
